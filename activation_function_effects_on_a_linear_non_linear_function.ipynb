# A snippet to plot and understand how activation functions (relu, sigmoid, softmax, leaky relu,tan h ) affect tan/sin/cos/exp (a non linear function) or linear spread.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate non-linear input data (tan/sin/cos/exp transformation)
x = np.linspace(-1.4, 1.4, 50)
x_nonlinear = np.sin(x)

# Define weight and bias
weight = 1.5
bias = 0.5

#compute pre-activation values
z = weight * x_nonlinear + bias  # Using x_nonlinear (use 'x' incase of linear)

#activation functions
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

def relu(z):
    return np.maximum(0, z)

def leaky_relu(z, alpha=0.09):
    return np.where(z > 0, z, alpha * z)

def softmax(z):
    exp_z = np.exp(z - np.max(z))  # Subtract max(z) for numerical stability
    return exp_z / np.sum(exp_z)



sigmoid_activation = sigmoid(z)
tanh_activation = tanh(z)
relu_activation = relu(z)
leaky_relu_activation = leaky_relu(z)
softmax_activation = softmax(z)



plt.figure(figsize=(10, 5))
plt.plot(x, z, 'o--', label="Pre-activation (Non-linear Output)")
plt.plot(x, sigmoid_activation, 's-', label="Sigmoid Activation")
plt.plot(x, tanh_activation, 'd-', label="Tanh Activation")
plt.plot(x, relu_activation, '^-', label="ReLU Activation")
plt.plot(x, leaky_relu_activation, 'x-', label="Leaky ReLU Activation")
plt.plot(x, softmax_activation, 'p-', label="Softmax Activation")


plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.title("Effect of Non-linear Input on Activation Functions")
plt.axhline(0, color="black", linewidth=0.5, linestyle="dotted")
plt.legend()
plt.grid()

#display plot
plt.show()
```
